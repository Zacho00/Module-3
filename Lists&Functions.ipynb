9.2 Define a generator function called get_odds() that returns the odd numbers from range(10). Use a for loop to find and print the third value returned.

```python
def get_odds():
    odds = []
    for number in range(10):
        odd = number % 2
        if odd >= 1:
            odds.append(number)
    print(odds[2])
get_odds()
```
